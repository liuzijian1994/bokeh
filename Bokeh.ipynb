```python
import numpy as np

from bokeh.layouts import gridplot
from bokeh.plotting import figure, show, output_file
from bokeh.sampledata.stocks import GOOG,AAPL
from bokeh.io import output_notebook
#bokeh.sampledata.download()
```

```python
def datetime(x):
    return np.array(x, dtype=np.datetime64)
```

```python
pic1 = figure(x_axis_type="datetime", title="Google Stock Closing Prices")
pic1.grid.grid_line_alpha=0.1
pic1.xaxis.axis_label = 'Date'
pic1.yaxis.axis_label = 'Close_Price'
```

```python
pic1.line(datetime(AAPL['date']), AAPL['adj_close'], color='#33A02C', legend='Apple')
pic1.line(datetime(GOOG['date']), GOOG['adj_close'], color='#FB9A99', legend='Google')

goog_price = np.array(GOOG['adj_close'])
goog_dates = np.array(GOOG['date'], dtype=np.datetime64)

window_size = 7
window = np.ones(window_size)/float(window_size)
goog_avg = np.convolve(goog_price, window, 'same')

pic2 = figure(x_axis_type="datetime", title="Google Weekly Average")
pic2.xaxis.axis_label = 'Date'
pic2.yaxis.axis_label = 'Avg_Price'

pic2.circle(goog_dates, goog_price, size=2, legend='close', alpha=0.2)

pic2.line(goog_dates, goog_avg, legend='avg', color='red')

output_notebook()

show(gridplot([[pic1,pic2]], plot_width=480, plot_height=360))  
```
